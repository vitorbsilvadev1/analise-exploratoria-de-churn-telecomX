```python
# Barplot - Quanto do lucro vem de cliente de alto risco vs seguro
# Objetivo: Demonstrar quanto ainda pode ser convertido em receita segura

faturamento_total_atual = df[df['churn'] == 'No']['gasto_total'].sum()

filtro_risco = (
    ((df['idoso'] == "Yes") | (df['parceiro'] == 'No') | (df['dependentes'] == 'No')) &
    (df['tipo_contrato'] == 'Month-to-month') &
    (df['metodo_pagamento'] == 'Electronic check') &
    (df['suporte_tecnico'] == 'No') &
    (df['seguranca_online'] == 'No')
)

faturamento_risco = df[filtro_risco]['gasto_total'].sum()
faturamento_seguro = faturamento_total_atual - faturamento_risco

dados_plot = pd.DataFrame({
    'Status': ['Receita em Risco', 'Receita Segura'],
    'Valor (R$)': [faturamento_risco, faturamento_seguro]
})

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

grafico = sns.barplot(x='Status', y='Valor (R$)', data=dados_plot, palette=['#e74c3c', '#2ecc71'])

for p in grafico.patches:
    grafico.annotate(f'R$ {p.get_height():,.2f}',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9),
                     textcoords = 'offset points',
                     fontsize=12, fontweight='bold')

plt.title('Lifetime Value \nClientes de Risco x Fidelizados', fontsize=15, pad=20)
plt.ylabel('Gasto Total do cliente (R$)')
plt.ylim(0, faturamento_total_atual * 1.1)

figbarplot_lucro = plt.gcf()

plt.show()
```

# Relatório Estratégico de Retenção e Análise de Churn – TelecomX

## 1. Introdução

O objetivo desta análise é identificar os padrões de comportamento que levam à evasão de clientes (*Churn*) na TelecomX. O problema central reside na perda de receita e na instabilidade do fluxo de caixa causada por cancelamentos prematuros. Esta análise visa transformar dados brutos em um plano de ação para aumentar o *Lifetime Value* (LTV) e garantir a saúde financeira da companhia.

## 2. Limpeza e Tratamento de Dados

Para assegurar a precisão dos insights, os dados passaram pelo seguinte processo de tratamento:

* **Normalização:** Conversão de dados brutos (JSON) para um formato tabular estruturado.
* **Renomeação:** Tradução das variáveis para o português (ex: de `MonthlyCharges` para `gasto_mensal`) para alinhar a análise técnica ao vocabulário de negócios.
* **Feature Engineering:** Criação da métrica de `gasto_diario` (calculada como `gasto_mensal / 30`) para avaliar a percepção de valor do cliente sob uma ótica de consumo cotidiano.
* **Limpeza:** Verificação de tipos de dados e tratamento de valores ausentes para evitar distorções estatísticas.

## 3. Análise Exploratória de Dados (EDA)

## 4. Conclusões e Insights

* **Fator de Blindagem:** Clientes de fibra ótica com serviços de suporte e segurança integrados podem cancelar significativamente menos dos que aqueles que não tem. A proteção técnica atua como uma prevenção de estresse, principalmente a quem não tem facilidade com tecnologia (como o público idoso).

* **Vulnerabilidade Operacional**: Identificamos que uma parcela significativa do faturamento provém de clientes com baixo vínculo contratual (mensal) e métodos de pagamento manuais. Converter esses perfis para modelos de fidelidade não é apenas uma medida defensiva, mas uma estratégia para maximizar o LTV (Lifetime Value), garantindo que a receita em risco se transforme em capital previsível para o crescimento da empresa.

## 5. Recomendações

Com base nas evidências, proponho as seguintes ações:

1. **Conversão para Plano Anual e crédito automático:** Implementar uma campanha focada no grupo de risco (Idosos ou Solitários com Cheque Eletrônico), oferecendo um desconto estratégico para migração ao contrato de 1 ano com pagamento via **Cartão de Crédito Automático**.

2. **Consolidação do Valor da Fibra** (Upgrade e Reformulação): Propomos a reformulação do portfólio de Fibra Ótica, transformando o Suporte Técnico e a Segurança Online em atributos nativos do serviço. Para a base atual que ainda não utiliza esses recursos, deve-se conceder um upgrade gratuito de fidelidade, enquanto para novos clientes, o plano já deve ser comercializado como uma "solução completa" de conectividade blindada. O objetivo é elevar o valor percebido e criar uma barreira de saída baseada em utilidade e conveniência, garantindo a manutenção da margem de lucro e a proteção do faturamento sem a necessidade de descontos agressivos no preço da fatura

3. **Régua de Boas-Vindas** (Acompanhamento Ativo): Como os dados mostram que o risco de cancelamento é altíssimo nos primeiros 6 meses, sugiro que a empresa não "esqueça" o cliente após a venda. Realizando um cronograma de contatos automáticos por e-mails ou mensagens para garantir que ele aprenda a usar o serviço e se sinta valorizado desde o primeiro dia, oferecendo ajuda, ensinando a usar o Suporte Técnico e garantindo que ele saiba que tem Segurança Online inclusa..
